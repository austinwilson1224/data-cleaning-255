```python
weather = pd.read_csv(path + "weather.csv")
weather.head()
```

- we want the element column to be column values 

```python
weather_melt = pd.melt(
    weather,
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp'
)
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```
